1. What is a random variable in probability theory?

ANS- A  random variable is a function that maps the outcomes of a random phenomenon to numerical values. It's not a variable in the traditional algebraic sense that takes on an unknown value, but rather a function that assigns a real number to each possible outcome in a sample space.

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

Ans- A probability distribution function (PDF) and a cumulative distribution function (CDF) describe a random variable in different but related ways.

A PDF describes how densely probability is distributed around different values of a continuous random variable. It does not give probability directly at a single point; instead, probabilities are found by calculating the area under the curve of the PDF over a range of values. The PDF is always non-negative, and the total area under its curve over all possible values is equal to 1.

A CDF, on the other hand, shows the total probability accumulated up to a certain value. It gives the probability that the random variable is less than or equal to a given value. Unlike the PDF, the CDF directly provides probability values. The CDF always starts at 0, increases as the variable increases, and approaches 1 as the value goes to infinity.
